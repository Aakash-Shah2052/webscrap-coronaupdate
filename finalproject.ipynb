```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url = "https://www.worldometers.info/coronavirus/#countries"
response = requests.get(url)
virus = response.text
virus = BeautifulSoup(virus,'lxml')
virus
```

```python
get_table = virus.find("table",id = "main_table_countries_today")
get_table_data = get_table.tbody.find_all("tr")
get_table_data
```

```python
get_table_data = get_table_data[8:50]
get_table_data
```

```python
list = []
for i in get_table_data:
    list.append(i.text)
print(list)    
```

```python
output = []
for i in range(len(list)):
    x = list[i].split('\n')
    y = x[1:6]
    output.append(y)
print(output)    
```

```python
x = ['#',
 'Country',
 'Total Cases',
 'New Cases',
 'Total Deaths',
 'New Deaths',
 'Total Recovered',
 'Active',
  'Serious Critical'  ]
y = []
for i in output:
  if i == x:
    continue
  y.append(i)
print(y)
```

```python
final_output = [x]+y
final_output
```

```python
import csv   
fields=final_output
with open('finalproject.csv', 'w') as f:
    x = csv.writer(f)
    for i in fields:
      x.writerow(i)
```
